```python
import numpy as np
import matplotlib.pyplot as plt

# 
x1 = np.array([0, 5000, 10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000])
y1= np.array([4.64e-5, 1e-5, 4.64e-6, 4.64e-7, 1e-7, 1e-7, 1e-7, 1e-7, 1e-7, 1e-7])
y2=np.array([1e-4, 2.15e-5, 2.15e-5, 2.15e-5, 1e-5, 1e-7, 1e-6, 1e-6, 1e-5, 2.15e-6])
y3=np.array([1e-4, 4.64e-6, 2.15e-6, 4.64e-6, 4.64e-6, 1e-6, 4.64e-7, 2.15e-5, 4.64e-7, 1e-6])
y4=np.array([4.64e-5, 4.64e-6, 2.15e-6, 4.64e-7, 1e-7, 1e-7, 1e-7, 1e-7, 1e-7, 1e-7])


for i in range(len(x1) - 1):
    plt.plot([x1[i], x1[i+1]], [y1[i], y1[i]], color='red')  # 
    plt.plot([x1[i], x1[i+1]], [y2[i], y2[i]], color='blue')
plt.plot([45000, 50000], [y1[9], y1[9]], color='red')
plt.plot([45000, 50000], [y2[9], y2[9]], color='blue')

for i in range(len(x1) - 1):
    plt.plot([x1[i], x1[i+1]], [y3[i], y3[i]*(0.999 ** 1250)], color='orange')  # 
    plt.plot([x1[i], x1[i+1]], [y4[i], y4[i]*(0.999 ** 1250)], color='green')
plt.plot([45000, 50000], [y3[9], y3[9]], color='orange')
plt.plot([45000, 50000], [y4[9], y4[9]], color='green')
# 
for i in range(1, len(x1)):
    plt.plot([x1[i], x1[i]], [y1[i-1], y1[i]], linestyle='dashed', color='red')
    plt.plot([x1[i], x1[i]], [y2[i-1], y2[i]], linestyle='dashed', color='blue')
    plt.plot([x1[i], x1[i]], [y3[i-1]*(0.999 ** 1250), y3[i]], linestyle='dashed', color='orange')
    plt.plot([x1[i], x1[i]], [y4[i-1]*(0.999 ** 1250), y4[i]], linestyle='dashed', color='green')
plt.scatter(x1, y1, color='red', marker='o', label="Points")
plt.scatter(x1, y2, color='blue', marker='o', label="Points")
plt.scatter(x1, y3, color='orange', marker='o', label="Points")
plt.scatter(x1, y4, color='green', marker='o', label="Points")
plt.yscale("log")
plt.xlabel("Data amount")
plt.ylabel("Learning Rate")
plt.grid()
plt.show()
```
