```python
import pandas as pd

# Excelファイルの読み込み
file_path = "data.xlsx"  # ファイルパスを指定
df = pd.read_excel(file_path)

df=df.iloc[4:369]
```

```python
df
```

```python
df=df[['Unnamed: 0','Unnamed: 1']]
```

```python
df
```

```python
# 条件に基づいて置換する関数を定義
def replace_weather(x):
    if x.startswith("晴"):
        return "sunny"
    elif x.startswith("曇"):
        return "cloudy"
    elif x.startswith("雨"):
        return "rainy"
    return "other"  # 上記条件に該当しない場合はそのまま返す

# 関数を適用
df["Unnamed: 1"] = df["Unnamed: 1"].apply(replace_weather)
# Unnamed: 0列を日付型に変換してフォーマットを変更
df["Unnamed: 0"] = pd.to_datetime(df["Unnamed: 0"]).dt.strftime("%-m/%-d")
# 結果を表示
print(df)

# 結果を表示
print(df)
df.to_excel("weatherdata.xlsx", index=False)
```
